```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

model = YOLO("/kaggle/working/grape_yolo11/grape_yolo11_l1002/weights/best.pt")

#testing on random image from internet
test_img = "/kaggle/input/botrytis-test2"

results = model.predict(source=test_img, imgsz=640, conf=0.25, show=False, save=True)

# Display result
res_plotted = results[0].plot()  
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(res_plotted, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/kaggle/working/grape_yolo11/grape_yolo11_l1002/results.csv")

# Plotting the mAP and losses
plt.figure(figsize=(10,6))
plt.plot(df["epoch"], df["metrics/mAP50(B)"], label="mAP@50")
plt.plot(df["epoch"], df["metrics/mAP50-95(B)"], label="mAP@50-95")
plt.xlabel("Epoch")
plt.ylabel("Score")
plt.title("Validation Performance")
plt.legend()
plt.show()
```

```python
import random
import supervision as sv
from ultralytics import YOLO

model = YOLO("/kaggle/working/grape_yolo11/grape_yolo11_l1002/weights/best.pt")

ds = sv.DetectionDataset.from_yolo(
    images_directory_path="/kaggle/input/grape-disease-merged-dataset/grape_disease_merged/images/test",
    annotations_directory_path="/kaggle/input/grape-disease-merged-dataset/grape_disease_merged/labels/test",
    data_yaml_path="/kaggle/input/data-yaml-file/data.yaml"
)


# Random sample
i = random.randint(0, len(ds) - 1)
image_path, image, target = ds[i]

results = model(image, verbose=False)[0]
detections = sv.Detections.from_ultralytics(results).with_nms()

box_annotator = sv.BoxAnnotator()
label_annotator = sv.LabelAnnotator()

annotated_image = image.copy()
annotated_image = box_annotator.annotate(scene=annotated_image, detections=detections)
annotated_image = label_annotator.annotate(scene=annotated_image, detections=detections)

sv.plot_image(annotated_image)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

base_path = "/kaggle/working"
files = [
    "/grape_yolo11/grape_yolo11_l1002/BoxP_curve.png",
    "/grape_yolo11/grape_yolo11_l1002/BoxR_curve.png",
    "/grape_yolo11/grape_yolo11_l1002/BoxPR_curve.png",
    "/grape_yolo11/grape_yolo11_l1002/BoxF1_curve.png"
]


titles = ["Precision Curve", "Recall Curve", "PR Curve", "F1 Curve"]

plt.figure(figsize=(12, 10))

for i, (file, title) in enumerate(zip(files, titles), 1):
    img = mpimg.imread(base_path + file)
    plt.subplot(2, 2, i)
    plt.imshow(img)
    plt.title(title)
    plt.axis("off")

plt.tight_layout()
plt.show()
```
